```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
```

```python
def triangle_wave(x, period=1, amplitude=2):
  return amplitude * abs(((x-period/4) % period) - period/2) / (period/2)-amplitude/2

t = np.linspace(0,3.49, 1000)
sine = np.sin(2*np.pi*t)
# Create square wave data
sq = np.array([1 if i % 2 == 0 else -1 for i in np.floor(2 * t)])
tri = triangle_wave(t)
fig, axs = plt.subplots(nrows=3, ncols=1, sharex=True, figsize=(8, 6))
axs[0].plot(t, sine, color='magenta')
axs[0].set_yticklabels([])
axs[0].set_xticklabels([])
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[0].hlines(0, t[-1], 0, color='black', linestyles='dashed')
axs[0].annotate(text='(a)', xy=(3.5, -1), xytext=(3.45, -0.9), fontsize=14)
axs[0].annotate(text=r'$V_{pp}$', xy=(0.6, 0.1), xytext=(0.55, 0.15))
axs[0].arrow(0.5, -1, 0, 1.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[0].arrow(0.5, 1, 0, -1.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[0].annotate(text=r'$V$', xy=(1.26, 0.1), xytext=(1.3, 0.15))
axs[0].arrow(1.25, 0, 0, 0.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[0].annotate(text=r'$T$', xy=(2.75, 1), xytext=(2.7, 0.75))
axs[0].arrow(2.25, 1, 0.9, 0, head_width=0.2, head_length=0.1, fc='k', ec='k',capstyle='round')
axs[0].arrow(3.25, 1, -0.9, 0, head_width=0.2, head_length=0.1, fc='k', ec='k',capstyle='round')
axs[1].plot(t, sq, color='blue')
axs[1].set_yticklabels([])
axs[1].set_xticklabels([])
axs[1].set_xticks([])
axs[1].set_yticks([])
axs[1].hlines(0, t[-1], 0, color='black', linestyles='dashed')
axs[1].annotate(text='(b)', xy=(3.5, -1), xytext=(3.45, -0.9), fontsize=14)
axs[1].annotate(text=r'$V_{pp}$', xy=(0.6, 0.1), xytext=(0.55, 0.15))
axs[1].arrow(0.5, -1, 0, 1.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[1].arrow(0.5, 1, 0, -1.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[1].annotate(text=r'$V$', xy=(1.26, 0.1), xytext=(1.3, 0.15))
axs[1].arrow(1.25, 0, 0, 0.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[1].annotate(text=r'$T$', xy=(2.75, 1), xytext=(2.7, 0.75))
axs[1].arrow(2.25, 1, 0.9, 0, head_width=0.2, head_length=0.1, fc='k', ec='k',capstyle='round')
axs[1].arrow(3.25, 1, -0.9, 0, head_width=0.2, head_length=0.1, fc='k', ec='k',capstyle='round')
axs[2].plot(t, tri, color='green')
axs[2].set_yticklabels([])
axs[2].set_xticklabels([])
axs[2].set_xticks([])
axs[2].set_yticks([])
axs[2].hlines(0, t[-1], 0, color='black', linestyles='dashed')
axs[2].annotate(text='(c)', xy=(3.5, -1), xytext=(3.45, -0.9), fontsize=14)
axs[2].annotate(text=r'$V_{pp}$', xy=(0.6, 0.1), xytext=(0.55, 0.15))
axs[2].arrow(0.5, -1, 0, 1.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[2].arrow(0.5, 1, 0, -1.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[2].annotate(text=r'$V$', xy=(1.26, 0.1), xytext=(1.3, 0.15))
axs[2].arrow(1.25, 0, 0, 0.8, head_width=0.1, head_length=0.2, fc='k', ec='k',capstyle='round')
axs[2].annotate(text=r'$T$', xy=(2.75, 1), xytext=(2.7, 0.75))
axs[2].arrow(2.25, 1, 0.9, 0, head_width=0.2, head_length=0.1, fc='k', ec='k',capstyle='round')
axs[2].arrow(3.25, 1, -0.9, 0, head_width=0.2, head_length=0.1, fc='k', ec='k',capstyle='round')
# Add a common x and y label for all subplots
fig.supxlabel("Time")
fig.supylabel("Voltage")
plt.savefig('waves.svg')
plt.show()
```

```python
t = np.linspace(0,3.49, 1000)
# Create square wave data
sq = np.array([1 if i % 2 == 0 else -1 for i in np.floor(2 * t)]) + 1
RC=np.zeros(1000)
for j in range(0,8,2):
    print(j)
    for i in range(144):
        if i+j*143 < 1000:
            RC[i+j*143] = 2*(1-np.exp(-15*(t[i])))
        if i+j*143+144 < 1000:
            RC[i+j*143+144] = 2*np.exp(-15*(t[i]))
fig, axs = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(8, 6))
axs[0].plot(t, sq, color='blue')
axs[0].annotate(text='(a)', xy=(3.5, 0), xytext=(3.45, 0.1), fontsize=14)
axs[0].set_yticklabels([])
axs[0].set_xticklabels([])
axs[0].set_xticks([])
axs[0].set_yticks([])

axs[1].plot(t, RC, color='red')
axs[1].annotate(text='(b)', xy=(3.5, 0), xytext=(3.45, 0.1), fontsize=14)
axs[1].set_yticklabels([])
axs[1].set_xticklabels([])
axs[1].set_xticks([])
axs[1].set_yticks([])

# Add a common x and y label for all subplots
fig.supxlabel("Time")
fig.supylabel("Voltage")
plt.savefig('RC.svg')
plt.show()
```

```python
tau = np.linspace(1.5, 1.5+1/15, 100)
h = np.ones(100)*2/2.7

plt.plot(t, sq, '-b', label=r'$V_o$')
plt.plot(t, RC, '-r', label=r'$V_C$')
plt.annotate(text=r'$V_o$', xy=(0.9,1.5), xytext=(0.82, 2))
plt.hlines(2, 0, 1.5, linestyles='dashed', color='black')
plt.annotate(text=r'$V_o/e$', xy=(0.9,.7), xytext=(0.8, 2/2.7))
plt.hlines(2/2.7, 0, 1.5+1/15, linestyles='dashed', color='black')
plt.annotate(text=r'$V=0$', xy=(0.9,0), xytext=(0.8, 0))
plt.hlines(0, 0, 3.5, linestyles='dashed', color='black')
plt.annotate(text=r'$\Delta t=1/RC$', xy=(1.51,0), xytext=(1.45, -0.15))
plt.vlines(1.5+1/15, 0, 2/2.7, linestyles='dashed', color='black')
plt.fill_between(tau, h, alpha=0.2, color='red')
plt.xlim(0.9, 1.99)
plt.ylim(-0.2, 2.25)
plt.legend()
plt.xticks([])
plt.yticks([])
#plt.xlabel('Time')
#plt.ylabel('Voltage')
plt.savefig('RCzoom.svg')
plt.show()
```

```python
x = np.linspace(0,1,100)
y = x
# Create a figure and axes
fig, ax = plt.subplots()
# Define the center of the arc
center = (0, 0)
# Define the radius of the arc
radius = 1
# Define the start and end angles of the arc in degrees
start_angle = 0
end_angle = 45
# Create the arc patch
arc = Arc(center, 0.5 * radius, 0.5 * radius, angle=0, theta1=start_angle, theta2=end_angle, edgecolor='k')

ax.fill_between(x, y, facecolor='yellow', edgecolor='black', linewidth=1, alpha=0.5)
ax.add_patch(arc)
ax.annotate(text=r'$\omega t$', xy=(0.3, 0.2), xytext=(0.25, 0.1), fontsize=14)
ax.annotate(text=r'$Z_{total}$', xy=(0.5, 0.5), xytext=(0.40, 0.55), fontsize=14)
ax.arrow(0.0, 0.0, 0.97, 0.97, head_width=0.05, head_length=0.05, fc='k', ec='k',capstyle='round')
ax.annotate(text=r'$X_C$', xy=(1, 0.5), xytext=(0.9, 0.45), fontsize=14)
ax.arrow(1, 0.0, 0, 0.92, head_width=0.05, head_length=0.05, fc='k', ec='k',capstyle='round')
ax.annotate(text=r'$X_R$', xy=(0.0, 0.0), xytext=(0.5, 0.05), fontsize=14)
ax.arrow(0.0, 0.0, 0.95, 0, head_width=0.05, head_length=0.05, fc='k', ec='k',capstyle='round')
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.set_xticks([])
ax.set_yticks([])
plt.xlabel('Real Impedance')
plt.ylabel('Imaginary Impedance')
plt.savefig('ImpedancePhase.svg')
plt.show()
```
